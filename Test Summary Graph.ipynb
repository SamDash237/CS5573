```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import openpyxl as px
%matplotlib inline
```

```python
results = px.load_workbook('testdata.xlsx')
resultSheet = results['Sheet1']
```

```python
ser = [ resultSheet['B%s'%i].value for i in range(2,7) ]
n1 = [ resultSheet['C%s'%i].value for i in range(2,7) ]
n2 = [ resultSheet['D%s'%i].value for i in range(2,7) ]
n4 = [ resultSheet['E%s'%i].value for i in range(2,7) ]
n6 = [ resultSheet['F%s'%i].value for i in range(2,7) ]
n8 = [ resultSheet['G%s'%i].value for i in range(2,7) ]
```

```python
barWidth = 0.15

rs = np.arange(len(ser))
r1 = [x + barWidth for x in rs]
r2 = [x + barWidth for x in r1]
r4 = [x + barWidth for x in r2]
r6 = [x + barWidth for x in r4]
r8 = [x + barWidth for x in r6]

plt.figure(figsize=(19.2,14.4))

plt.bar(rs, ser, color='red', width=barWidth, edgecolor='white', label='Serial')
plt.bar(r1, n1, color='orange', width=barWidth, edgecolor='white', label='Parallel N=1')
plt.bar(r2, n2, color='green', width=barWidth, edgecolor='white', label='Parallel N=2')
plt.bar(r4, n4, color='blue', width=barWidth, edgecolor='white', label='Parallel N=4')
plt.bar(r6, n6, color='cyan', width=barWidth, edgecolor='white', label='Parallel N=6')
plt.bar(r8, n8, color='purple', width=barWidth, edgecolor='white', label='Parallel N=8')

plt.xlabel('Tests', fontweight='bold', fontsize=18)
plt.ylabel('Time (seconds)', fontweight='bold', fontsize=18)
plt.xticks([r + barWidth for r in range(len(ser))], ['1', '2', '3', '4', 'Average'], fontsize = 16)
plt.yticks(fontsize = 16)
plt.title("Testing Results", fontweight='bold', fontsize=24)

plt.legend(fontsize= 16)
plt.show()
```
